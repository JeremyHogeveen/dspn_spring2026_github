## 2.2 Why should we care?

* <b>Big Data</b>: the amopunt of data that you can now collect and/or analyze has grown substantially in the last few decades.

* <b>The Curse of Dimensionality</b>: As measurement technologies and/or modes of delivery improve, the dimensionality of data is also increasing. Highly multivariate and complex data is the new norm, and this can often defy our expectations and intuitions.

* <b>Reproducibility</b>: Increasing expectations for open data and open code, so researchers anywhere can recreate, and perhaps expand upon your work.

* <b>Ethical Implications</b>: Large psych and neuro datasets can shape decision-making in clinical practice and across society—interpreting data correctly is important.
