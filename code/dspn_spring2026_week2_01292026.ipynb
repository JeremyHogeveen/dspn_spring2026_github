# Section 1: Computational Environments

## 1.1 What do you mean by "environment"?

* Data science involves different pieces of software that do different things
  * Big consideration for Python "libraries".
  * Lots of libraries we may use in Python. We then likely need to use some different ones for different projects.
  * How can we manage project-specific sets of Python libraries without breaking everything?
* A computational envrionment is a directory on your computer that contains all of the software dependencies used in one project.
  * I recommend using Conda/Anaconda package manger to help with this!

## 1.2 Installing Anaconda

<img src="img/anaconda.png" width=400>

* Anaconda will help us to create and maintain a computing "environment" that should have all the packages we need for the course.
* Any time we want to add new functions, we can activate the environment—we will call "dspn_env"—and add them that way.
* We will install it from the command line now that you have some confidence there!
  * If you want to install using standard GUI approach, <a href="https://www.anaconda.com/docs/getting-started/anaconda/install">here is documentation on this</a>
  
1. If Mac:
* Open your Terminal
* run: <i>curl -O https://repo.anaconda.com/archive/Anaconda3-2025.12-2-MacOSX-arm64.sh</i>

2. If Windows:
* Open Git bash
* run: <i>curl https://repo.anaconda.com/archive/Anaconda3-2025.12-1-Windows-x86_64.exe --output .\Downloads\Anaconda3-2025.12-1-Windows-x86_64.exe</i>
* Navigate to Downloads folder in Windows Explorer, double click Anaconda installer
  * Install with default settings, should work great!
* Once installed, search Windows for "Anaconda Prompt". Open...
  *  run: <i>conda init bash</i>
  *  This is an important step. It will initialize conda in your "bash_profile," which will be seen when you open git bash! (i.e., will help git bash and conda  talk to each other)

## 1.3 Setting up the DSPN course environment with anaconda

1. Open git bash (if Windows) or Terminal (if Mac)
2. Create the conda environment:
* Run: <i>conda create -n dspn_env python=3.10 spyder jupyterlab ipykernel numpy pandas matplotlib seaborn scikit-learn statsmodels -y</i>

3. Activate the dspn_env:
* Run: <i>conda activate dspn_env</i>

4. Add a dspn_env to Jupyter Notebook or JupyterLab so we can select it as a "Kernel" when running code.
* Run: <i>python -m ipykernel install --user --name=dspn_env --display-name "Python (dspn_env)"</i>

5. Move to your cloned <i>dspn_spring2026_github</i> directory

6. Launch Jupyter Lab!

# Section 2: Doing some coding in Python

## 2.1 Writing Python code in Jupyter

* Python is an __object-oriented__ programming language
  * The things you create and/or read into Python are invariably some kind of __object__
* Python objects can be, ___variables___

```python
variable = 6
print(variable)
print('your first variable equals:',variable)
```

* Variables are most often __integers, floats, or strings__

```python
floating_point_numbers = 3.14
integer = 3
string = 'pi'
print(string,'is equal to:',floating_point_numbers,'which is approximately:',integer)
```

* Multiple variables can be assigned at once in Python

```python
multiple, thing, line = floating_point_numbers, integer, string
print(multiple)
print(thing)
print(line)
```

* Mathematical operators can be applied to numerical variables
  * Order of operations follows B() E** D/ M* A+ S-

```python
variable_2 = 14
math_answer = (variable_2-variable)**2/30
print('your first mathematical solution is:', math_answer)
```

* Division gives a few other options...
  * if you want the closest integer, and it's remainder you can do...

```python
math_answer_2 = (variable_2-variable)**2 // 30 # // to get a rounded integer of the solution
remainder = (variable_2-variable)**2 % 30 # % to get the remainder
print('the closest integer is',math_answer_2,', and its remainder is',remainder)
```

* Python objects might also be collections (e.g. data frames and lists)

```python
list_1 = [-96,-66,-87,86,-88,-25,24,59,-99,59]
print('here is your first list :',list_1)
```
